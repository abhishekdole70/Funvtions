```python
def calculator(operation:str,n1:float,n2:float):
    if operation=='add':
        c=n1+n2
    elif operation=='subtract':
        c=n1-n2
    elif operation=='multiply':
        c=n1*n2
    elif operation=='divide':
        c=n1/n2
    else:
        raise Exception('Only add,subtract,divide and multiple options are available')
    return c
```

```python
calculator('multiply',2,4)
```

```python
def aot(a,b,c):
    s=(a+b+c)/2
    A=(s*(s-a)*(s-b)*(s-c))**(1/2)
    print(f'Area of Triangle having sides {a}, {b} and {c} is {A}')
```

```python
aot(2,3,4)
```

```python
def fib(n):
    a,b=0,1
    for i in range(n+1):
        yield a
        a,b =b, a+b
```
